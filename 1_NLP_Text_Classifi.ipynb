```python
# Tokenzation -->Stop words-->Stemming-->Lemmatization
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
# Tokenzation---> Convert paragraph sentenses to words
nltk.download('punkt')
nltk.download('punkt_tab')
sen = nltk.sent_tokenize(Paragraph)
```

```python
#Stemming
stemmer=PorterStemmer()
print(stemmer.stem("Thinking"))
print(stemmer.stem("Drinking"))
print(stemmer.stem("History"))
```

```python
#Lemmatization

import nltk
nltk.download('stopwords') # Download stopwords resource

# The original code iterated over characters instead of words.
# It should iterate over the tokenized words.
# Also, typically stopwords are checked against lowercase versions of words.

for text_item in corpus:
  words = nltk.word_tokenize(text_item)
  filtered_and_stemmed_words = []
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary= True)
x=cv.fit_transform(corpus)
```

```python
## Apply stop words
import re
corpus=[]
for i in range(len(sen)):
  review=re.sub('[^a-zA-Z]',' ',sen[i])
  review=review.lower()
  review=review.split()
  review=[stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
  corpus.append(" ".join(review)) # Join the list of words back into a single string
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary= True)
```

```python
x=cv.fit_transform(corpus)
cv.vocabulary_
```
